```python
def read_sr_files(file_name):

    with open(file_name, 'r') as file:
        line_list = file.readlines()
    depth = []
    gamma = []
    for i, line in enumerate(line_list):
        if  i == 0:
            continue
        number = float(line)
        if i % 2 != 0:
            depth.append(number)
        if i % 2 == 0:
            gamma.append(number)
        else:
            continue
    well_df = pd.DataFrame({"Depth": depth, "Gamma": gamma})
    return well_df

file_name = r"C:\Users\gabri\OneDrive\Desktop\PhD\Python for Geo\HACKATHON\blockhead\INEL1GA"
file_name1 = r"C:\Users\gabri\OneDrive\Desktop\PhD\Python for Geo\HACKATHON\blockhead\INEL1GB1"
file_name2 = r"C:\Users\gabri\OneDrive\Desktop\PhD\Python for Geo\HACKATHON\blockhead\INEL1GB2"
file_name3= r"C:\Users\gabri\OneDrive\Desktop\PhD\Python for Geo\HACKATHON\blockhead\INEL1GB3"
file_name4= r"C:\Users\gabri\OneDrive\Desktop\PhD\Python for Geo\HACKATHON\blockhead\INEL1GB4"
file_name5= r"C:\Users\gabri\OneDrive\Desktop\PhD\Python for Geo\HACKATHON\blockhead\INEL1GC"

well_df = read_sr_files(file_name)
well_df

well_df1 = read_sr_files(file_name1)
well_df1

well_df2 = read_sr_files(file_name2)
well_df2

well_df3 = read_sr_files(file_name3)
well_df3

well_df4 = read_sr_files(file_name4)
well_df4

well_df5 = read_sr_files(file_name5)
well_df5
```

### Real data example

Synthetics are not real ... here we provide a collection of timeseries from a well log. This kind of data is characterized by *fairly* abrupt changes to the physical rock properties sensed by the tools as they pass through different geologic formations. The challenge in this real world example is that there are no perfect hard edges in the series.

The data satyr_3_wire_ow.las is provided by Geoscience Australia and use under a CC-BY 4.0 license. Additional data associated with this well can be found on the [NOPIMS website](https://nopims.dmp.wa.gov.au/NOPIMS/Search/WellDetails?id=ENO0595344#).

```python
scale_space = create_scale_space(gamma_ray, 
                                 min_scale=5, 
                                 max_scale=2000,
                                 num_scales=100)

contours = measure.find_contours(scale_space["second"], 0)
interval_tree = populate_interval_tree(contours)

scale_space1= create_scale_space(gamma_ray1, 
                                 min_scale=5, 
                                 max_scale=2000,
                                 num_scales=100)

contours1 = measure.find_contours(scale_space1["second"], 0)
interval_tree1 = populate_interval_tree(contours1)

scale_space2= create_scale_space(gamma_ray2, 
                                 min_scale=5, 
                                 max_scale=2000,
                                 num_scales=100)

contours2 = measure.find_contours(scale_space2["second"], 0)
interval_tree2 = populate_interval_tree(contours2)

scale_space3= create_scale_space(gamma_ray3, 
                                 min_scale=5, 
                                 max_scale=2000,
                                 num_scales=100)

contours3 = measure.find_contours(scale_space3["second"], 0)
interval_tree3 = populate_interval_tree(contours3)

scale_space4= create_scale_space(gamma_ray4, 
                                 min_scale=5, 
                                 max_scale=2000,
                                 num_scales=100)

contours4 = measure.find_contours(scale_space4["second"], 0)
interval_tree4 = populate_interval_tree(contours4)

scale_space5= create_scale_space(gamma_ray5, 
                                 min_scale=5, 
                                 max_scale=2000,
                                 num_scales=100)

contours5 = measure.find_contours(scale_space5["second"], 0)
interval_tree5 = populate_interval_tree(contours5)
```

```python
import matplotlib.pyplot as plt
gamma_ray_list = [gamma_ray, gamma_ray1,gamma_ray2, gamma_ray3, gamma_ray4, gamma_ray5]
scale_space_list = [scale_space, scale_space1, scale_space2, scale_space3, scale_space4, scale_space5]
interval_tree_list = [interval_tree, interval_tree1, interval_tree2, interval_tree3, interval_tree4, interval_tree5]

fig, axes = plt.subplots(len(gamma_ray_list), 1, figsize=(10, len(gamma_ray_list) * 4))
for i, (gamma_ray_data, scale_space_data, interval_tree_data) in enumerate(zip(gamma_ray_list, scale_space_list, interval_tree_list)):
    # Use the default_display for each dataset and plot it on its own axis
    plt.sca(axes[i])
    fig = default_display(gamma_ray_data, scale_space_data, interval_tree_data)
    plt.title(f'Dataset {i + 1}')

plt.tight_layout()
plt.show()
```

```python
from blockhead.util import harmonic_average, slice_by_scale
gamma_ray_list = [gamma_ray1, gamma_ray2, gamma_ray3, gamma_ray4, gamma_ray5]
interval_tree_list = [interval_tree1, interval_tree2, interval_tree3, interval_tree4, interval_tree5]
#  set an arbitrary threshold, to define the scale to "block" the well at
threshold = 60
results_list = []
for i, (gamma_ray_data, interval_tree_data) in enumerate(zip(gamma_ray_list, interval_tree_list)):
    results = slice_by_scale(interval_tree_data['root'], threshold, gamma_ray_data, fn=harmonic_average)
    results_list.append(results)
```

```python


plt.figure(figsize=(8,15))
for i, (gamma_ray_data, results) in enumerate(zip(gamma_ray_list, results_list)):
    plt.subplot(1, len(gamma_ray_list), i + 1)  # Create subplots for each dataset
    plt.plot(gamma_ray_data, range(len(gamma_ray_data)), label=f'Dataset {i + 1}')


    for result in results:
        node, data = result
        xmin, xmax = node.top_edge, node.bottom_edge
        ymin, ymax = data, data
        plt.plot([ymin, ymax], [xmin, xmax], '-r')
    plt.ylim([len(gamma_ray), 0])

    plt.xlabel('Gamma Ray')
    plt.ylabel('Depth')
    plt.title(f'Dataset {i + 1}')
    plt.legend()

# Adjust layout and show the plot
plt.tight_layout()
plt.show()
```
